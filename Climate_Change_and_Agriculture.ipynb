```python
Correlations = df[["Year" , "Average_Temperature_C" , "Total_Precipitation_mm" , "CO2_Emissions_MT" , "Crop_Yield_MT_per_HA" ,
                   "Extreme_Weather_Events" , "Irrigation_Access_%" , "Pesticide_Use_KG_per_HA" , "Fertilizer_Use_KG_per_HA" ,
                   "Soil_Health_Index" , "Economic_Impact_Million_USD" ]].corr()
plt.figure(figsize=(10,9))
sns.heatmap( Correlations ,annot = True , cmap = 'Purples')
plt.title("Correlations")
```
